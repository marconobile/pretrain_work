```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Path to the .pth file
pth_file_path = "/storage_common/nobilm/pretrain_paper/guacamol/500k_experiment/TEST_new_env_relu_exp_post_merge/last_model.pth"

try:
    # Load the file
    data = torch.load(pth_file_path, map_location=torch.device('cpu'))

    print("Contents of the .pth file:")
    if isinstance(data, dict):
        # Iterate and print keys/summary if it's a state_dict or data dict
        for key, value in data.items():
            if isinstance(value, torch.Tensor):
                print(f"{key}: Tensor of shape {value.shape}")
                # Visualize image-like tensors (assuming shape [C, H, W])
                if value.ndimension() == 3 and value.shape[0] in [1, 3]:
                    img = value.numpy().transpose(1, 2, 0)
                    img = img.squeeze() if img.shape[-1] == 1 else img
                    plt.imshow(img, cmap='gray' if img.ndim == 2 else None)
                    plt.title(f"Visualization of {key}")
                    plt.show()
            else:
                print(f"{key}: {type(value)}")
    else:
        print("The content is not a dictionary.")
        print(data)

except FileNotFoundError:
    print(f"File not found: {pth_file_path}")
except Exception as e:
    print(f"Error loading the .pth file: {e}")
```
